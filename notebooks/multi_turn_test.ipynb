```python
import openai
import pandas as pd
import os
from IPython.display import display

# Load OpenAI API Key
openai.api_key = os.getenv("OPENAI_API_KEY")

#Models to test
models = {
    "GPT-4o (OpenAI API)": {"name": "gpt-4o", "type": "api"},
    "GPT-4o-mini (OpenAI API)": {"name": "gpt-4o-mini", "type": "api"},
    "o1-mini (OpenAI API)": {"name": "o1-mini", "type": "api"},
}

# Test scenarios
test_conversations = [
    {
        "test_name": "Engagement & Relevance",
        "turns": [
            "Heyy! How’s your day going?",
            "That's nice! Tell me about something exciting you did this week?",
            "Ooh, that’s cool! Do you have anything planned for the weekend?",
        ]
    },
    {
        "test_name": "Memory & Context Retention",
        "turns": [
            "I'm planning a trip to Japan thi summer. Should I visit Tokyo or Kyoto?",
            "Kyoto sounds nice in my opinion. What are some things I can't miss while there?",
            "I love history! Should visit Nara too?",
        ]
    },
    {
        "test_name": "Adaptability & Subject Change",
        "turns": [
            "I’m feeling really sad right now.",
            "Nevermind I'm just overreacting? Scratch that, tell me something funny.",
            "Haha, thanks! Okay, now what’s your most inspiring quote?",
        ]
    },
    {
        "test_name": "Contradiction & Self-Checking",
        "turns": [
            "What’s your favorite color?",
            "Stop lying, you told me it was green last time?",
            "Are you sure? Maybe you forgot or something, I don't know.",
        ]
    }
]

# Store results
multi_turn_results = []

# Generate responses
def generate_response(model_name, messages):
    """Generates a multi-turn chat from a given model."""
    model_info = models[model_name]
    
    client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))  # Load API key
    response = client.chat.completions.create(
        model=model_info["name"],
        messages=messages
    )
    return response.choices[0].message.content

# Run tests on all models
for model_name in models.keys():
    for test in test_conversations:
        test_name = test["test_name"]
        conversation_log = []
        messages = [{"role": "user", "content": test["turns"][0]}]  # First user prompt

        # Run n_turns rounds
        for i in range(len(test["turns"])):
            bot_response = generate_response(model_name, messages)
            conversation_log.append({
                "Turn": i + 1,
                "Test Name": test_name,
                "Model": model_name,
                "User Prompt": messages[-1]["content"],
                "Bot Response": bot_response
            })
            
            if i + 1 < len(test["turns"]):  # If there are more turns left
                messages.append({"role": "assistant", "content": bot_response})
                messages.append({"role": "user", "content": test["turns"][i + 1]})

        multi_turn_results.extend(conversation_log)

# Convert results
df_multi_turn = pd.DataFrame(multi_turn_results)

# Display results
display(df_multi_turn)

# Export results
csv_filename = "multi_turn_response_comparison1.csv"
df_multi_turn.to_csv(csv_filename, index=False)
print(f"Results saved to {csv_filename}")
```
